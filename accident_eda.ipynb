On what hour of the day does most accidents occur

```python
import seaborn as sns
```

```python
df.Start_Time.dt.hour
```

```python
sns.histplot(df.Start_Time.dt.hour, bins= 24)
```

What day of the week does the highest number of accidents occur

```python
sns.histplot(df.Start_Time.dt.day_of_week, bins=7)
```

compare accident hour of sunday and monday with the help of a graph

```python
sun = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.histplot(sun.dt.hour, bins= 24)

```

```python
sunday = df[df.Start_Time.dt.dayofweek == 6]
sns.histplot(sunday.Start_Time.dt.hour, bins=24)
```

```python
monday = df[df.Start_Time.dt.dayofweek == 0]
sns.histplot(monday.Start_Time.dt.hour, bins=24)
```

Which month has the highest accident occurence

```python
sns.histplot(df.Start_Time.dt.month, bins= 12)
```

Analyse accident distribution for a particular year

```python
year_wise = df[df.Start_Time.dt.year == 2019] 
sns.histplot(year_wise.Start_Time.dt.month, bins= 12)
```
